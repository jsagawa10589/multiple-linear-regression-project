```python
#This dataset analyzes 50 business startups in different states
#The variables in the dataset include R&D spending, Administration, Marketing spending, State & Profit
#We will create a multiple linear regression model to predict profit based on some of the variables mentioned above

import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline

#Read in the dataset
data = pd.read_csv('/Users/jermainesagawa/Downloads/50_Startups.csv') 

#Displays the data
data.head()

#Lets plot each category vs. Profit to see their linear relationship

plt.scatter(data["R&D Spend"], data.Profit)
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.show()

plt.scatter(data["Administration"], data.Profit)
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.show()

plt.scatter(data["Marketing Spend"], data.Profit)
plt.xlabel("Marketing Spend")
plt.ylabel("Profit")
plt.show()

plt.scatter(data["State"], data.Profit)
plt.xlabel("State")
plt.ylabel("Profit")
plt.show()

#Based on our results, it is evident that R&D Spend and Marketing Spend have a better linear correlation with Profit
#We will use the above two categories as our X (independent variables)
#We will set Profit as our y (dependent variable)

#We set these two columns to X using data.iloc[:, :[0,2]]
#Python indexes start from 0, so we use 0 for the first column R&D Spend, and 2 for the third column Marketing Spend
X = data.iloc[:, [0,2]]

#To set our y we will use data.iloc[:, :4], basically take all rows and only the last column 
y = data.iloc[:, [4]]

#Now we split our data to train and test with a 80:20 ratio respectively 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#Using our training set we create a model 
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

#With the model we can predict y_ using the testing set
y_predict = regressor.predict(X_test)

#We now evaluate our model
from sklearn.metrics import r2_score

#compute RSquared
r2 = r2_score(y_test, y_predict)
print("the R2 score is: ", r2)

#Based on our results, our model can make some good predictions however there is room for improvement.






```
